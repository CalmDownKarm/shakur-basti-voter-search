```python
import geopandas as gpd
```

```python
toilets = gpd.read_file('gejsons/Toilets.json')
```

```python
toilets.head()
```

```python
toilets.info()
```

```python
toilets.drop('geometry', axis=1).to_sql('toilets', engine, if_exists='replace', index=None)
```

```python
engine.table_names()
```

```python
night_shelters = gpd.read_file('geojsons/NightShelters.json')
```

```python
night_shelters.head()
```

```python
agency_names = {agency:agency for agency in night_shelters.agency_nam.unique()}
agency_names.update({
    'Centre for Equity Studies':'Centre For Equity Studies',
    'Rachna Womens Development Society':'Rachna Womenâ€™s Development Society'
})
night_shelters.rename(columns={'Y':'lat', 'X':'lon', 'Rev_Distt':'rev_dist', 'Division': 'divison', 'agency_nam':'agency_name'})\
    .drop('geometry', axis=1)\
    .assign(agency_name = lambda x: x.agency_name.map(agency_names)).to_sql('night_shelters', engine, index=None, if_exists='replace')
```

```python
engine.table_names()
```

```python
pd.read_sql('select * from night_shelters', engine).groupby('rev_dist').count()
```
